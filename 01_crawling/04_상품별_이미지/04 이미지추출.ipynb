```python
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains

from bs4 import BeautifulSoup
import os
import numpy as np
import pandas as pd
import requests
import json
import urllib.request
import re
import time
import datetime
```

```python
def create_folder(folder_path):
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(f"폴더가 생성되었습니다: {folder_path}")
        time.sleep(1)
        
    else:
        print(f"이미 폴더가 존재합니다: {folder_path}")
        time.sleep(1)

# Url/경로 주면 이미지 다운받는 함수
def download_image(url, save_path):
    response = requests.get(url)
    response.raise_for_status()
    with open(save_path, 'wb') as file:
        file.write(response.content)
        
# 이미지 추출함수
def get_img(funding_id):
    url = f'https://www.wadiz.kr/web/campaign/detail/{funding_id}?_refer_section_st=PREORDER_1'
    driver.get(url)
    time.sleep(3)
    soup = BeautifulSoup(driver.page_source,"html.parser")
    k = 0
    img_url = []
    
    count = len(soup.find_all("div",{"class":"inner-contents fr-view"})[0].find_all("img"))

    folder_path = f"D:/wadiz/{category_folder}/{funding_id}"   ###################
    create_folder(folder_path)

    for i in range(1,count):
        temp = str(funding_id) + "_" + str(k)
        file_url = soup.find_all("div",{"class":"inner-contents fr-view"})[0].find_all("img")[i].get("data-src")
        print(file_url)

        if ("png" in file_url):

            save_path = f"D:/wadiz/{category_folder}/{funding_id}/{temp}(png).png"   ###################
            download_image(file_url, save_path)
            k+=1
            img_url.append([funding_id,file_url])
            time.sleep(1)

        elif ("img" in file_url):
            
            save_path = f"D:/wadiz/{category_folder}/{funding_id}/{temp}(img).img"   ###################
            download_image(file_url, save_path)
            k+=1
            img_url.append([funding_id,file_url])
            time.sleep(1)

        elif ("gif" in file_url):
            
            save_path = f"D:/wadiz/{category_folder}/{funding_id}/{temp}(gif).gif"   ###################
            download_image(file_url, save_path)
            k+=1
            img_url.append([funding_id,file_url])
            time.sleep(1)

        elif ("jpg" in file_url):
            
            save_path = f"D:/wadiz/{category_folder}/{funding_id}/{temp}(jpg).jpg"   ###################
            download_image(file_url, save_path)
            k+=1
            img_url.append([funding_id,file_url])
            time.sleep(1)

        elif ("jpeg" in file_url):
            
            save_path = f"D:/wadiz/{category_folder}/{funding_id}/{temp}(jpeg).jpeg"   ###################
            download_image(file_url, save_path)
            k+=1
            img_url.append([funding_id,file_url])
            time.sleep(1)

            
        elif ("bmp" in file_url):
            
            save_path = f"D:/wadiz/{category_folder}/{funding_id}/{temp}(bmp).bmp"   ###################
            download_image(file_url, save_path)
            k+=1
            img_url.append([funding_id,file_url])
            time.sleep(1)
            
            
    return img_url


# 와디즈 프로젝트 주소 https://www.wadiz.kr/web/campaign/detail/208241

```

```python
code = list(category_id['상품번호'])
a = code[0].replace("'",'"')          #################### code = 카테고리 0~17
```

```python
category = list(category_id['카테고리'])
category_id = json.loads(a)

print(category)
```

```python
category_folder = category[0]          #################### code = 카테고리 0~17

print(category_folder)
```

```python
# category_id에 상품 List code 담으면 이미지 추출
for funding_id in category_id:
    try:
        temp = get_img(funding_id)
    except:
        pass
    print(temp)
    if temp == funding_id:
        fail_list.append(temp)

    else:
        img_url.append(temp)

# 실패한 list 정리
fail_list_pd = pd.DataFrame(fail_list)
fail_list_pd.to_excel('fail_list.xlsx')

# 성공한 list + url 정리
img_url_pd = pd.DataFrame(img_url)
img_url_pd.to_excel('img_url_list.xlsx')

print("complete!!")
```
